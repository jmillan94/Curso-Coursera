```python
#Rellenar NA

df_filling=pd.DataFrame({'A':[43,59,np.nan,6,3,np.nan],
'B':[1,20,np.nan,43,np.nan,32],'C':[np.nan,32,6,4,np.nan,2]})
df_filling
```

```python
#Eliminarlos por filas axis=0 y por columnas axis=1

df_filling.dropna(axis=0)
```

```python
#Rellenarlos por algun valor

df_filling.fillna(value='paila', inplace=True)
```

```python
df_filling
```

```python
#Rellenar por la media o mediana
df_filling.fillna(value=df_filling["A"].median())
```

```python
df_filling
```

```python
#Rellenar por el valor de atras
df_filling.fillna(method='ffill')
#Rellenar por el valor de adelante
df_filling.fillna(method='bfill')
```
